```python
dataset.drop(columns=['SK_ID_CURR'], inplace=True) #ID columns irrelevant
dataset.drop(columns=['FLAG_MOBIL'], inplace=True) #FLAG_MOBIL: All values are same
```

```python
dataset.drop(columns=['OWN_CAR_AGE'], inplace=True) # OWN_CAR_AGE null percent more than 60 %
```

```python
#Feature engineering
dataset['DAYS_EMPLOYED'] = dataset['DAYS_EMPLOYED'].apply(lambda x: -x if x > 0 else x)
```

```python
print(dataset['DAYS_BIRTH'].max())
print(dataset['DAYS_EMPLOYED'].max())
print(dataset['DAYS_REGISTRATION'].max())
print(dataset['DAYS_ID_PUBLISH'].max())
```

```python
dataset['Age'] = ((dataset['DAYS_BIRTH'] * (-1)) / 365).astype(int)
dataset.drop(columns='DAYS_BIRTH', inplace=True)

dataset['EMPLOYED_DURATION'] = ((dataset['DAYS_EMPLOYED'] * (-1)) / 365).astype(int)
dataset.drop(columns='DAYS_EMPLOYED', inplace=True)

dataset['REGISTRATION_DURATION'] = ((dataset['DAYS_REGISTRATION'] * (-1)) / 365).astype(int)
dataset.drop(columns='DAYS_REGISTRATION', inplace=True)

dataset['ID_PUBLISH_DURATION'] = ((dataset['DAYS_ID_PUBLISH'] * (-1)) / 365).astype(int)
dataset.drop(columns='DAYS_ID_PUBLISH', inplace=True)

```

```python
dataset.isnull().sum()
```

```python
#Impute null numbers

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')

dataset['AMT_ANNUITY'] = imputer.fit_transform(dataset[['AMT_ANNUITY']]).ravel()
dataset['AMT_GOODS_PRICE'] = imputer.fit_transform(dataset[['AMT_GOODS_PRICE']]).ravel()
dataset['NAME_TYPE_SUITE'] = imputer.fit_transform(dataset[['NAME_TYPE_SUITE']]).ravel()
dataset['CNT_FAM_MEMBERS'] = imputer.fit_transform(dataset[['CNT_FAM_MEMBERS']]).ravel()
dataset['OCCUPATION_TYPE'] = imputer.fit_transform(dataset[['OCCUPATION_TYPE']]).ravel()

```

```python
dataset.isnull().sum()
```

```python
import pandas as pd
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

numerical_columns = dataset.select_dtypes(include=['float64', 'int64']).columns
valid_numerical_columns = [col for col in numerical_columns if dataset[col].nunique() > 2]

dataset[valid_numerical_columns] = scaler.fit_transform(dataset[valid_numerical_columns])
print(valid_numerical_columns)
```

```python
dataset = dataset.drop_duplicates()
```

```python
dataset.duplicated().sum()
```

```python
#Encode process for columns

from sklearn.preprocessing import LabelEncoder

categorical_columns = dataset.select_dtypes(include=['object']).columns
binary_columns = categorical_columns[(dataset[categorical_columns].nunique() == 2) & (categorical_columns != 'Target')]


label_encoders = {}

for column in binary_columns:
  le = LabelEncoder()
  dataset[column] = le.fit_transform(dataset[column])
  label_encoders[column] = le  #

print(binary_columns)
```

```python
import pandas as pd

import pandas as pd

categorical_columns = dataset.select_dtypes(include=['object']).columns

non_binary_columns = categorical_columns[(dataset[categorical_columns].nunique() > 2)]

dataset = pd.get_dummies(dataset, columns=non_binary_columns, drop_first=False)

print(non_binary_columns)


```

```python
# Step 6: PCA (retain 95% variance)
X = dataset.drop(['TARGET'], axis=1)
y = dataset['TARGET']
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95, random_state=42)
X_pca = pca.fit_transform(X)
```

```python
# Step 7: Upsample minority class
df_combined = pd.concat([pd.DataFrame(X_pca), y.reset_index(drop=True)], axis=1)
df_majority = df_combined[df_combined['TARGET'] == 0]
df_minority = df_combined[df_combined['TARGET'] == 1]
```

```python
from sklearn.utils import resample


df_minority_upsampled = resample(df_minority,
                                 replace=True,
                                 n_samples=len(df_majority),
                                 random_state=42)
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
X_balanced = df_upsampled.drop('TARGET', axis=1)
y_balanced = df_upsampled['TARGET']
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.2, random_state=42)
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
# Step 9: Model Training
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Step 10: Evaluation
y_pred = rf_model.predict(X_test)
report = classification_report(y_test, y_pred, output_dict=True)
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
print(report)
```

```python
from sklearn.metrics import accuracy_score

# Reduce the number of trees for faster training
rf_model = RandomForestClassifier(n_estimators=50, random_state=42)
rf_model.fit(X_train, y_train)

# Predict and evaluate
y_pred = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, output_dict=True)
conf_matrix = confusion_matrix(y_test, y_pred)

accuracy
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression # Import LogisticRegression

# Ensemble (Voting) Classifier without SVC due to memory issue
# Rebuild ensemble with Random Forest and Logistic Regression only

# Create and train the Logistic Regression model
log_model = LogisticRegression(random_state=42) # Initialize log_model
log_model.fit(X_train, y_train) # Train log_model


voting_clf_simple = VotingClassifier(
    estimators=[
        ('rf', rf_model),
        ('log', log_model)
    ],
    voting='soft'
)

# Train ensemble model
voting_clf_simple.fit(X_train, y_train)

# Predict and evaluate
y_pred_ensemble_simple = voting_clf_simple.predict(X_test)
accuracy_ensemble_simple = accuracy_score(y_test, y_pred_ensemble_simple)
report_ensemble_simple = classification_report(y_test, y_pred_ensemble_simple, output_dict=True)
conf_matrix_ensemble_simple = confusion_matrix(y_test, y_pred_ensemble_simple)

accuracy_ensemble_simple
```
